```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
font = FontProperties(fname="SimHei.ttf", size=14)
```

```python
df = pd.read_csv('不同交易者的4个阶段.csv')
df['SSE_change'] = df['SSE_change'] + 1
df.set_index('交易日期', drop=True, inplace=True)
df
```

```python
df_bg_1 = df[df['知情交易者交易活跃度_stage'] == 1]
df_bg_2 = df[df['知情交易者交易活跃度_stage'] == 2]
df_bg_3 = df[df['知情交易者交易活跃度_stage'] == 3]
df_bg_4 = df[df['知情交易者交易活跃度_stage'] == 4]
print('知情交易者不同阶段收益率：')
print('升温：', float(np.cumprod(df_bg_1['SSE_change'])[-1:]))
print('活跃：', float(np.cumprod(df_bg_2['SSE_change'])[-1:]))
print('冷却：', float(np.cumprod(df_bg_3['SSE_change'])[-1:]))
print('低迷：', float(np.cumprod(df_bg_4['SSE_change'])[-1:]))

plt.subplot(2, 2, 1)
plt.plot(np.cumprod(df_bg_1['SSE_change']))
plt.xticks(df_bg_1['SSE_change'].index[::40], rotation=45)
plt.title('stage 1')

plt.subplot(2, 2, 2)
plt.plot(np.cumprod(df_bg_2['SSE_change']))
plt.xticks(df_bg_2['SSE_change'].index[::40], rotation=45)
plt.title('stage 2')

plt.subplot(2, 2, 3)
plt.plot(np.cumprod(df_bg_3['SSE_change']))
plt.xticks(df_bg_3['SSE_change'].index[::40], rotation=45)
plt.title('stage 3')

plt.subplot(2, 2, 4)
plt.plot(np.cumprod(df_bg_4['SSE_change']))
plt.xticks(df_bg_4['SSE_change'].index[::40], rotation=45)
plt.title('stage 4')

plt.tight_layout()

plt.show()
```
